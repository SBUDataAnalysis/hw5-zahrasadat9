## sam2

```python
stack=np.stack((sam2.x,sam2.y), axis = 0)
cov= np.cov(stack)
print('sam1 cov:','\n',cov)

landa, eighen=np.linalg.eig(cov)

teta= np.arccos(eighen[0][0])

print('new cov: ',np.cov(eighen))

cos= np.cos(teta)
sin= np.sin(teta)
sam11=sam2.to_numpy()*np.array([[cos,sin],[-sin,cos]])[:, None]
x1=np.sum(sam11[0],axis=1)
y1=np.sum(sam11[1],axis=1)

plt.hist2d(x1,y1,bins=100);
```
